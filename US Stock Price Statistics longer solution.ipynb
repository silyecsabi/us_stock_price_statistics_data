```python
import pandas as pd
df = pd.read_csv('us_stock_historical_data.csv')
```

```python
df.drop('Volume', axis = 1, inplace = True)    # irrelevant column
```

```python
df.rename(columns = {'Close/Last' : 'Close'}, inplace = True)
```

```python
df.set_index('Date', inplace = True)
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df['weekly_high'] = df[['High']].resample('w-mon', closed = 'left').max().resample('d').ffill()
```

```python
df[40:80]
```

```python
q_values = df.resample('Q').max()
q_values.drop('Close', axis = 1, inplace = True)
q_values.drop('Open', axis = 1, inplace = True)
q_values.drop('Low', axis = 1, inplace = True)
q_values.rename(columns = {'High' : 'Quarter Max High'}, inplace = True)
```

```python
q_min = df.resample('Q').min()
q_min.drop('Close', axis = 1, inplace = True)
q_min.drop('Open', axis = 1, inplace = True)
q_min.drop('High', axis = 1, inplace = True)

q_min_list = q_min['Low'].tolist()
```

```python
q_values.insert(loc = 1, column = 'Quarter Min Low', value = q_min_list)
q_values = q_values[::-1]
```

```python
m_values = df.resample('M').max()
m_values.drop('Close', axis = 1, inplace = True)
m_values.drop('Open', axis = 1, inplace = True)
m_values.drop('Low', axis = 1, inplace = True)
m_values.rename(columns = {'High' : 'Monthly Max High'}, inplace = True)
```

```python
m_min = df.resample('M').min()
m_min.drop('Close', axis = 1, inplace = True)
m_min.drop('Open', axis = 1, inplace = True)
m_min.drop('High', axis = 1, inplace = True)

m_min_list = m_min['Low'].tolist()
```

```python
m_values.insert(loc = 1, column = 'Monthly Min Low', value = m_min_list)
m_values = m_values[::-1]
```

```python
w_values = df.resample('W').max()
w_values.drop('Close', axis = 1, inplace = True)
w_values.drop('Open', axis = 1, inplace = True)
w_values.drop('Low', axis = 1, inplace = True)
w_values.rename(columns = {'High' : 'Weekly Max High'}, inplace = True)
```

```python
w_min = df.resample('W').min()
w_min.drop('Close', axis = 1, inplace = True)
w_min.drop('Open', axis = 1, inplace = True)
w_min.drop('High', axis = 1, inplace = True)

w_min_list = w_min['Low'].tolist()
```

```python
w_values.insert(loc = 1, column = 'Weekly Min Low', value = w_min_list)
w_values = w_values[::-1]
```

```python
df.insert(loc = 4, column = 'Weekly Max', value = None)
df.insert(loc = 5, column = 'Weekly Min', value = None)
df.insert(loc = 6, column = 'Monthly Max', value = None)
df.insert(loc = 7, column = 'Monthly Min', value = None)
df.insert(loc = 8, column = 'Quarter Max', value = None)
df.insert(loc = 9, column = 'Quarter Min', value = None)
```

```python
nearest_locs = []
for date_to_find in q_values.index:
    nearest_locs.append(df.index.get_loc(date_to_find, method = 'nearest'))
```

```python
for i in range (len(nearest_locs)):
    df.at[df.index[nearest_locs[i]], 'Quarter Max'] = q_values.loc[q_values.index[i]]['Quarter Max High']
    df.at[df.index[nearest_locs[i]], 'Quarter Min'] = q_values.loc[q_values.index[i]]['Quarter Min Low']
```

```python
nearest_locs = []
for date_to_find in m_values.index:
    nearest_locs.append(df.index.get_loc(date_to_find, method = 'nearest'))
```

```python
for i in range (len(nearest_locs)):
    df.at[df.index[nearest_locs[i]], 'Monthly Max'] = m_values.loc[m_values.index[i]]['Monthly Max High']
    df.at[df.index[nearest_locs[i]], 'Monthly Min'] = m_values.loc[m_values.index[i]]['Monthly Min Low']
```

```python
nearest_locs = []
for date_to_find in w_values.index:
    nearest_locs.append(df.index.get_loc(date_to_find, method = 'nearest'))
```

```python
for i in range (len(nearest_locs)):
    df.at[df.index[nearest_locs[i]], 'Weekly Max'] = w_values.loc[w_values.index[i]]['Weekly Max High']
    df.at[df.index[nearest_locs[i]], 'Weekly Min'] = w_values.loc[w_values.index[i]]['Weekly Min Low']
```
